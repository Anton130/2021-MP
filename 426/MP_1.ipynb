1.Найдите значение функции в точках 1, 5, 10, 100
$$ y = \log_{2x}|\frac{e^{cos(x)^A}tan(x^2)}{\frac{A}{9}x +\sqrt{x^3}}| $$
Где в качестве A подставьте дату (день) своего рождения.

```python
import numpy as np
import math 
def Func(x):
    A=30
    g=((math.exp(math.cos(x)**A))*(math.sin(x**2)/math.cos(x**2)))
    f=((A/9)*x)+x**(3/2)
    func=abs(g/f)
    print(np.log(func)/np.log(2*x))
Func(1)
Func(5)
Func(10)
Func(100)
```
